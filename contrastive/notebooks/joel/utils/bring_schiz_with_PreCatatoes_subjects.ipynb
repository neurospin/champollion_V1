```python
import pandas as pd
```

```python
used_schiz_file = "/neurospin/dico/data/deep_folding/current/datasets/schiz/aymeric_stratification/with_sub/used_schiz_subjects.csv"
precatatoes_file = "/neurospin/dico/data/bv_databases/human/partially_labeled/orbital_patterns/PreCatatoes/OFC_sulcal_type_data_186-subjects.csv"
precatatoes_file_columns_renamed = "/neurospin/dico/data/bv_databases/human/partially_labeled/orbital_patterns/PreCatatoes/OFC_sulcal_type_data_186-subjects_columns-renamed.csv"
precatatoes_file_only_schiz_and_control = "/neurospin/dico/data/bv_databases/human/partially_labeled/orbital_patterns/PreCatatoes/OFC_sulcal_type_data_186-subjects_only-schiz-and-control.csv"
target_file = "/neurospin/dico/data/deep_folding/current/datasets/schiz_extended/used_schiz_subjects.csv"
```

```python
schiz = pd.read_csv(used_schiz_file)
precatatoes = pd.read_csv(precatatoes_file)
```

```python
schiz.head()
```

```python
precatatoes.head()
```

```python
precatatoes = precatatoes.rename(columns={"N_anonymisation": "participant_id",
                                          "sexe": "sex",
                                          "pathologie": "diagnosis"})
```

```python
precatatoes
```

```python
precatatoes['sex'] = precatatoes['sex'].replace({"male": "M", "female": "F"})
precatatoes['diagnosis'] = precatatoes['diagnosis'].replace({"schizophrenie": "schizophrenia",
                                                             "schizophrenie ": "schizophrenia",
                                                             "sain": "control"})
precatatoes['participant_id'] = precatatoes['participant_id'].astype(str)
```

```python
precatatoes
```

```python
precatatoes.to_csv(precatatoes_file_columns_renamed, index=False)
```

```python
precatatoes = precatatoes[(precatatoes.diagnosis=="schizophrenia") | (precatatoes.diagnosis=="control")]
```

```python
precatatoes.to_csv(precatatoes_file_only_schiz_and_control, index=False)
```
