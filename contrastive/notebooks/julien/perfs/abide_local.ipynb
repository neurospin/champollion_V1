```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the results
#results = pd.read_csv('/neurospin/dico/data/deep_folding/current/models/Champollion_V1_after_ablation_latent_256/pred_asd_local_ndims256.csv')
results = pd.read_csv('/neurospin/dico/data/deep_folding/current/models/Champollion_V1_after_ablation/pred_asd_local_ndims32.csv')

results.sort_values(by='best_score', ascending=False, inplace=True)
results.columns = ['model', 'best_params', 'best_score', 'roc_auc_train_val', 'roc_auc_test']
# reformat index column
models = results['model']
results['model'] = [elem.split('/')[-4] for elem in models]
```

```python
# plot the correlation between train/val and test scores
plt.figure(figsize=(8, 8))
plt.scatter(results['roc_auc_train_val'], results['roc_auc_test'], alpha=0.5)
plt.xlabel('Train/Val ROC AUC')
plt.ylabel('Test ROC AUC')
plt.title('Train/Val vs Test ROC AUC')
```
