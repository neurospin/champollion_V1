```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
true_labels = pd.read_csv('/neurospin/dico/data/deep_folding/current/datasets/hcp/hcp_OFC_labels.csv', usecols=['Subject', 'Left_OFC'])
```

```python
# let's consider orbital with 4 labels
#proba1 = pd.read_csv('/volatile/jl277509/Runs/02_STS_babies/Output/2024-09-03/11-37-13_114/troiani_custom_embeddings/Left_OFC/test_probas_pred_0.csv')
proba1 = pd.read_csv('/volatile/jl277509/Runs/02_STS_babies/Program/Output/ORBITAL_12-layer_k7/20-56-02_2/troiani_custom_embeddings/Left_OFC/test_probas_pred_0.csv')
proba2 = pd.read_csv('/volatile/jl277509/Runs/02_STS_babies/Program/Output/ORBITAL_12-layer_k7/20-56-02_0/troiani_custom_embeddings/Left_OFC/test_probas_pred_0.csv')
```

```python
probas = pd.merge(left=proba1, right=proba2, on='Subject')
probas = pd.merge(left=probas, right=true_labels, on='Subject')
```

```python
fig, ax = plt.subplots(figsize=(4*5, 5), ncols=4)
for idx, (label1, label2) in enumerate(zip(['1', '2', '3', '4'], ['1_', '2_', '3_', '4_'])):
    ax[idx].scatter(probas[label1], probas[label2], c=probas['Left_OFC'])
    ax[idx].set_ylabel(f'label : {label1}')
plt.legend()
plt.show()
```
