```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sulcus = 'F.I.P.'
side = 'L'
skels = np.load(f'/neurospin/dico/data/deep_folding/current/datasets/UkBioBank/crops/2mm/{sulcus}/mask/{side}skeleton.npy')
```

```python
total = np.sum(skels!=0)
total_size = np.prod(skels.shape[1:])

list_prop_size = [1]
list_prop_vx = [1]
for k in range(10):
    skels_cropped = skels[:,k+1:-k-1, k+1:-k-1, k+1:-k-1, :]
    list_prop_size.append(np.prod(skels_cropped.shape[1:]) / total_size)
    list_prop_vx.append(np.sum(skels_cropped!=0) / total)
```

```python
plt.plot(range(0, len(list_prop_size)), list_prop_size, label='all voxels')
plt.plot(range(0, len(list_prop_vx)), list_prop_vx, label='Non zero voxels')
plt.xlabel('Distance to edge cropped')
plt.ylabel('Proportion of voxels')
plt.title(f'Sulcus : {sulcus}, side: {side}')
plt.legend()
plt.show()
```
