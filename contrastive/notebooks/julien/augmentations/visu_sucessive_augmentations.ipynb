```python
import anatomist.api as ana
from soma.qt_gui.qtThread import QtThreadCall
from soma.qt_gui.qt_backend import Qt
import dico_toolbox as dtx

from soma import aims
import numpy as np
import os
from tqdm import tqdm
```

```python
a = ana.Anatomist()
```

```python
dir = '/volatile2/jl277509/visu_augmentations/subs/'
imdirs = os.listdir(dir)
#cutout = 'cutin'
cutout = 'cutout'

for imdir in tqdm(imdirs):
    imdir = os.path.join(dir,imdir)

    skel0 = np.load(os.path.join(imdir, 'full_skel.npy'))
    skel1 = np.load(os.path.join(imdir, 'skel_trimdepth.npy'))
    skel2 = np.load(os.path.join(imdir, 'skel_trimdepth_extremities.npy'))
    skel_bottoms_before = (skel0 == 30).astype(np.int16)
    skel_bottoms_after = (skel2 == 30).astype(np.int16)
    skel3 = np.load(os.path.join(imdir, f'skel_trimdepth_extremities_{cutout}.npy'))

    skel0 = (skel0 !=0).astype(np.int16) # to be plot in red
    skel1 = (skel1 !=0).astype(np.int16)
    skel2 = (skel2 !=0).astype(np.int16)
    skel3 = (skel3 !=0).astype(np.int16)

    plot1 = skel0 - skel1 # removed vxs after 1st augm
    plot2 = skel0 - skel2 # after 2nd augm
    plot3 = skel0 - skel3 # after 3rd augm
    plot4 = skel3

    print(np.min(plot1)==0, np.min(plot2)==0, np.min(plot3)==0)
    plot0 = skel0.copy()
    print(f'directory : {imdir}')
    print(f'Total vx before: {np.sum(plot0)}')
    print((f'Total vx removed after augm1 : {np.sum(plot1)}'))
    print((f'Total vx removed after augm2 : {np.sum(plot2)}'))
    print((f'Total vx removed after augm3 : {np.sum(plot3)}'))
    print(f'Total vx after: {np.sum(skel3)}')
    # check that there are no vx which belong to skel_bottoms_after, but not to plot4
    print(np.sum(plot4))
    print(np.sum(np.logical_or(plot4, skel_bottoms_after))) # both numbers should be equal if keeping the bottoms

    for i, plot in enumerate([plot0, plot1, plot2, plot3, plot4, skel_bottoms_before, skel_bottoms_after]):
        im = aims.Volume(plot)
        bck = dtx.convert.volume_to_bucketMap_aims(im)
        aims.write(bck, os.path.join(imdir, f'bucket{i}.bck'))
```
