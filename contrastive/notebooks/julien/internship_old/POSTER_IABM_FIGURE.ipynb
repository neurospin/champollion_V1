```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import stats
import pandas as pd
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(20,4))

mu = 12
variance=16
r = 5
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + r*sigma, 100)
plt.fill(x, stats.norm.pdf(x, mu, sigma), c='black', alpha=0.8)

mu = 0
variance = 8
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + r*sigma, 100)
plt.fill(x, stats.norm.pdf(x, mu, sigma), c='darkred', alpha=0.8)

mu = 12
variance=16
r2 = 1
sigma = math.sqrt(variance)
x = np.linspace(mu + r2*sigma, mu + r*sigma, 100)
x_ = np.hstack((mu + r2*sigma - ((r2-r)*sigma/100),x))
y = stats.norm.pdf(x, mu, sigma)
y_ = np.hstack((np.array([0]),y))
plt.fill(x_, y_, c='forestgreen', alpha=0.8)

plt.savefig('/home/jl277509/Documents/IABM_2024_gaussian.png')
plt.show()
```
