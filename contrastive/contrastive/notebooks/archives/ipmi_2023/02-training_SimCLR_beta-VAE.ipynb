This notebook contains the code to load the latent spaces of beta-VAE and SimCLR

```python
import pandas as pd
```

```python
src_bVAE = "/neurospin/dico/data/deep_folding/papers/ipmi2023/models/beta-VAE/n_10/#5/cingulate_ACCpatterns_1_embeddings/full_embeddings.csv"
src_SimCLR = "/neurospin/dico/data/deep_folding/papers/ipmi2023/models/contrastive/trained_on_HCP_half_2/all/foldlabel/convnet-10/09-21-24_4/cingulate_ACCpatterns_1_embeddings/full_embeddings.csv"
dst = "/neurospin/dico/data/deep_folding/papers/ipmi2023/models/fusion/#5/cingulate_ACCpatterns_1_embeddings/full_embeddings.csv"
```

```python
bVAE = pd.read_csv(src_bVAE)
SimCLR = pd.read_csv(src_SimCLR)
```

```python
fusion = pd.merge(bVAE, SimCLR, left_on='ID', right_on='ID')
```
