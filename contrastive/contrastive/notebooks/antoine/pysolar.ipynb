```python
def run_shell_command_with_tty(command):
    try:
        result = subprocess.run(f'script -q -c "{command}" /dev/null', shell=True, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
        return result.stdout, result.stderr
    except subprocess.CalledProcessError as e:
        return e.stdout, e.stderr
    
def run_shell_command_with_pexpect(command):
    child = pexpect.spawn('/bin/bash', ['-c', command])
    child.logfile = open("/tmp/mylog", "wb")
    child.expect(pexpect.EOF)
    child.close()
    with open("/tmp/mylog", "rb") as f:
        output = f.read().decode()
    return output
```
