```python
import os
import numpy as np
import pandas as pd
from itertools import product
```

```python
## need to make sure that by default all probas are set to 0.
## manually, if cutin and cutout are in the same config, change the probas to 0.4.

cutin = {'1': ['proba_cutin=0.8', 'keep_proba_per_branch_cutin=0.5', 'patch_size_cutin=[20,90]']}
cutout = {'1': ['proba_cutout=0.8', 'keep_proba_per_branch_cutout=0.5', 'patch_size_cutout=[5,30]']}
translation = {'1': ['proba_translation=1', 'max_translation=1'],
               '2': ['proba_translation=1', 'max_translation=3'],
               '3': ['proba_translation=1', 'max_translation=5']}
rotation = {'1': ['proba_rotation=1', 'max_angle=12'],
            '2': ['proba_rotation=1', 'max_angle=18'],
            '3': ['proba_rotation=1', 'max_angle=30']}

augmentations = [cutin, cutout, translation, rotation]
```

```python
# Prepare options for each dictionary, including a None option
options = [[(i, v) for (k, v) in d.items()] + [None] for i, d in enumerate(augmentations)]

# Generate combinations and filter out invalid ones
combinations = [
    [item for item in combo if item is not None]
    for combo in product(*options)
    if (any(item is not None for item in combo))
]

list_configs = []
list_names = []

# list the configs
for combo in combinations:
    if len(combo)>1:
        # convert combo to list of list
        combo = [tup[1] for tup in combo]
        #Flatten and concatenate
        config = " ".join([item for sublist in combo for item in sublist])
        print(config)
        list_configs.append(config)

        name_config = config.replace("=", '_')
        name_config = name_config.replace(",", '_')
        name_config = name_config.replace(".", ' ')
        name_config = name_config.replace("[", ' ')
        name_config = name_config.replace("]", ' ')
        name_config = name_config.replace(" ", '_')
        print(name_config)
        list_names.append(name_config)
```

```python
## write to csv, then use in bash to generate the configs
a = pd.DataFrame(np.vstack((list_configs, list_names)).T)
a.to_csv('/neurospin/dico/jlaval/Output/ablation_2_models_full/configs_combination.csv', index=False, header=None, sep=';')
```
