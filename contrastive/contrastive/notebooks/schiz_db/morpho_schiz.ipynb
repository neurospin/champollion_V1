# Fusion des csv pour schiz

/!\ Cette fusion ne prend pas en compte de manière sûre les sessions. La première session qui arrive dans le csv est automatiquement choisie, alors que ce n'est peut-être pas la même que pour les squelettes.

```python
bases = ['bsnip1', 'candi', 'cnp', 'schizconnect-vip-prague']
sulcus = 'F.C.M.ant'
side = 'right'
```

```python
# import morpho csv
morphos = []

for base in bases:
    csv_path = f"/neurospin/dico/data/deep_folding/current/datasets/{base}/morphometry_statistics/{side}/morpho_{sulcus}._right.dat"
    morpho_base = pd.read_table(csv_path, sep='\s+')
    morphos.append(morpho_base)

full_morpho = pd.concat(morphos, axis=0, ignore_index=True)

full_morpho
```

```python
# once again, removing the sub
real_subject_names = list(full_morpho.subject)
for i,name in enumerate(real_subject_names):
    real_subject_names[i] = name[4:]
print(real_subject_names[:5])
full_morpho['subject'] = real_subject_names
full_morpho = full_morpho.sort_values(by='subject', ignore_index=True)
full_morpho
```

```python
# filter with the actual subjects used for schiz fusion

used_schiz_subjects = pd.read_csv("/neurospin/dico/data/deep_folding/current/datasets/schiz/used_schiz_subjects.csv")
used_schiz_subjects = used_schiz_subjects.sort_values(by='participant_id', ignore_index=True)
used_schiz_subjects
```

```python
# only keep subjects actually used in schiz
copied = full_morpho[full_morpho.subject.isin(used_schiz_subjects.participant_id)]
copied = copied.reset_index(drop=True)
print(copied.shape[0])

# remove subjects with valid != 1
copied = copied[copied.valid == 1]
print(copied.shape[0])
```

```python
# remove the second session with multiple sessions => /!\ not sure it is the right one
copied = copied[~copied.duplicated(subset='subject', keep='last')]
copied = copied.reset_index(drop=True)

# add diagnosis to copied
copied['diagnosis'] = used_schiz_subjects[used_schiz_subjects.participant_id.isin(copied.subject)].diagnosis.values

copied
```

```python
# check diagnosis is correct
for id in copied.subject.values:
    diag1 = used_schiz_subjects[used_schiz_subjects.participant_id == id].diagnosis.values
    diag2 = copied[copied.subject == id].diagnosis.values
    print(id, diag1, diag2)

    if diag1 != diag2:
        raise ValueError("Là")
```

```python
# save the df
copied.to_csv("/neurospin/dico/data/deep_folding/current/datasets/schiz/morphometry_statistics/F.C.M.ant/possibly_incorrect_right.csv")
```

# Extraction et calcul des paramètres pertinents

```python
morpho_df = pd.read_csv("/neurospin/dico/data/deep_folding/current/datasets/schiz/morphometry_statistics/F.C.M.ant/possibly_incorrect_right.csv", index_col=0)
morpho_df
```

```python
subject_parameters = ['subject']

labels = ['diagnosis']

shape_parameters = ['surface', 'connectedComponents', 'connectedComponentsAllRels', 'hullJunctionsLength',
                    'geodesicDepthMax', 'geodesicDepthMean', 'hullJunctionsLength']
position_parameters = ['gravityCenter_x', 'gravityCenter_y', 'gravityCenter_z', 'normal_x',
                       'normal_y', 'normal_z', 'direction_x', 'direction_y', 'direction_z']
```

```python
labels_df = morpho_df[subject_parameters + labels]

shape_df = morpho_df[subject_parameters + shape_parameters]
print(shape_df.head())
position_df = morpho_df[subject_parameters + position_parameters]
position_df.head()
```

```python
shape_df = shape_df.set_axis(['ID']+['dim_'+str(i) for i in range(1,shape_df.shape[1])], axis=1)
shape_df.to_csv("/neurospin/dico/data/deep_folding/current/datasets/schiz/morphometry_statistics/F.C.M.ant/shape_embs.csv", index=False)

shape_df
```

```python
position_df = position_df.set_axis(['ID']+['dim_'+str(i) for i in range(1,position_df.shape[1])], axis=1)
position_df.to_csv("/neurospin/dico/data/deep_folding/current/datasets/schiz/morphometry_statistics/F.C.M.ant/position_embs.csv", index=False)

position_df
```
