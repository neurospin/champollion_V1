```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------
# Data from Table 7  Occipital region (mean values)
# Order of methods: BT, SimCLR, Euclidean, Mahalanobis
# -------------------------------------------------
targets = ["log_Thickness", "log_surface_hull", "log_surface_total"]
methods = ["BT", "SimCLR", "Euclidean", "Mahalanobis"]

R2 = np.array([
    [0.0032, 0.0024, 0.0023, 0.0045],  # log_Thickness
    [0.2813, 0.2981, 0.3167, 0.4630],  # log_surface_hull
    [0.2731, 0.2916, 0.3109, 0.5375],  # log_surface_total
])

RMSE = np.array([
    [0.0881, 0.0882, 0.0882, 0.0881],
    [0.0258, 0.0255, 0.0252, 0.0223],
    [0.0323, 0.0319, 0.0315, 0.0258],
])

r_vals = np.array([
    [0.0584, 0.0531, 0.0522, 0.0684],
    [0.5305, 0.5461, 0.5629, 0.6805],
    [0.5228, 0.5401, 0.5577, 0.7332],
])

# -------------------------------------------------
# A simple, consistent color for each method
# -------------------------------------------------
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]  # BT, SimCLR, Euclidean, Mahalanobis

def plot_grouped_bars(mat, ylabel, title, filename, ylim=None, annotate=True):
    """
    mat: shape (n_targets, n_methods)
    """
    n_targets, n_methods = mat.shape
    x = np.arange(n_targets)
    width = 0.18

    fig, ax = plt.subplots(figsize=(8.5, 4.8))
    for m in range(n_methods):
        ax.bar(x + (m - (n_methods-1)/2)*width, mat[:, m], width,
               label=methods[m], color=colors[m])

        # Optional value labels on bars (nice for slides)
        if annotate:
            for i, v in enumerate(mat[:, m]):
                ax.text(x[i] + (m - (n_methods-1)/2)*width, v,
                        f"{v:.3f}" if v < 1 else f"{v:.4f}",
                        ha="center", va="bottom", fontsize=8, rotation=0)

    ax.set_xticks(x)
    ax.set_xticklabels(targets, rotation=12, ha="right")
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.legend(ncol=4, frameon=False)
    ax.grid(axis="y", linestyle="--", alpha=0.35)
    if ylim is not None:
        ax.set_ylim(*ylim)

    fig.tight_layout()
    fig.savefig(filename, dpi=200)
    print(f"Saved: {filename}")

# ---------------------------
# Make the three figures
# ---------------------------
# R^2  two ranges: very small for thickness, larger for surfaces
plot_grouped_bars(
    R2,
    ylabel=r"$R^2$",
    title="Occipital region  Regression performance (R)",
    filename="occipital_R2.png",
    ylim=(0.0, 0.60)  # shows the spread clearly
)

# RMSE  lower is better; choose a range that shows differences
plot_grouped_bars(
    RMSE,
    ylabel="RMSE",
    title="Occipital region  Regression performance (RMSE)",
    filename="occipital_RMSE.png",
    ylim=(0.020, 0.092)  # zoom around observed values
)

# r (correlation)  shows clear separation for surfaces
plot_grouped_bars(
    r_vals,
    ylabel="r",
    title="Occipital region  Regression performance (r)",
    filename="occipital_r.png",
    ylim=(0.0, 0.80)
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Data from Table  Occipital region (R2 with std, without log_Thickness)
# -----------------------------
targets = ["log_surface_hull", "log_surface_total"]
methods = ["BT", "SimCLR", "Euclidean", "Mahalanobis"]

# mean R2 values
r2_means = np.array([
    [0.2813, 0.2981, 0.3167, 0.4630],  # log_surface_hull
    [0.2731, 0.2916, 0.3109, 0.5375],  # log_surface_total
])

# std R2 values
r2_stds = np.array([
    [0.0258, 0.0255, 0.0252, 0.0223],  # log_surface_hull
    [0.0323, 0.0255, 0.0315, 0.0091],  # log_surface_total
])

# -----------------------------
# Colors for methods
# -----------------------------
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]  # BT, SimCLR, Euclidean, Mahalanobis

# -----------------------------
# Plot grouped bar chart with error bars
# -----------------------------
x = np.arange(len(targets))
width = 0.18

fig, ax = plt.subplots(figsize=(7.5, 4.5))
for m in range(len(methods)):
    ax.bar(x + (m - (len(methods)-1)/2)*width, r2_means[:, m], width,
           yerr=r2_stds[:, m], capsize=4,
           label=methods[m], color=colors[m], alpha=0.85)
    # Add value labels
    for i, v in enumerate(r2_means[:, m]):
        ax.text(x[i] + (m - (len(methods)-1)/2)*width, v + r2_stds[i, m] + 0.01,
                f"{v:.3f}", ha="center", va="bottom", fontsize=8)

ax.set_xticks(x)
ax.set_xticklabels(targets, rotation=12, ha="right")
ax.set_ylabel("$R^2$ (mean  std)")
ax.set_title("Occipital region  Regression performance ($R^2$)")
ax.legend(ncol=4, frameon=False)
ax.grid(axis="y", linestyle="--", alpha=0.35)
ax.set_ylim(0, 0.65)  # zoom to highlight differences

fig.tight_layout()
fig.savefig("occipital_r2.png", dpi=200)

print("Saved: occipital_r2.png")
```
